.apply() ‚Üí 'Go inside the column and modify each cell'

.assign() ‚Üí 'Add or update a column based on some logic'


üß† Difference Between .apply() and .assign() in Pandas

üîπ .apply() ‚Üí Works inside a column

Used to apply a function to each value in a Series (column) or row.

It transforms individual cells/values.

Usually used with lambda, regex, or math operations.
ex:
df['Name'] = df['Name'].apply(lambda x: re.sub('Raj', 'Ratnajit', x))


üîπ .assign() ‚Üí Works at the DataFrame level

Used to add or update entire columns based on logic.

Always returns a new DataFrame (does not modify the original unless reassigned).

Often combined with .apply() to build new or cleaned columns.
ex:
df = df.assign(Updated_Name=df['Name'].apply(lambda x: re.sub('Raj', 'Ratnajit', x)))


| Purpose                                  | Use                                      |
| ---------------------------------------- | ---------------------------------------- |
| Modify values inside a column            | `.apply()`                               |
| Add or update a column                   | `.assign()`                              |
| Combine transformation + column creation | `.assign(column = df['col'].apply(...))` |

üß† Understanding lambda x: in apply()

When you use df.apply(lambda x: ..., axis=1):

üëâ x represents one complete row of the DataFrame (as a Pandas Series).

The function runs once per row, so you can access individual columns inside that row using x['column_name'].
In simple words .apply() will loop through each row (since axis=1).

üìò Example:
df.apply(lambda x: x['Name'].upper(), axis=1)


Here, x is one row at a time.

x['Name'] gives the Name value from that specific row.

üîç Example Illustration:

For a DataFrame:

Name	Math_Scores	Gender
Raj	98	Male
Reshmi	35	Female

The lambda function runs like this:

For the first row ‚Üí
x = {'Name': 'Raj', 'Math_Scores': 98, 'Gender': 'Male'}

For the second row ‚Üí
x = {'Name': 'Reshmi', 'Math_Scores': 35, 'Gender': 'Female'}

‚úÖ In short:
When using apply(..., axis=1),
‚û°Ô∏è x = one row
When using apply(..., axis=0),
‚û°Ô∏è x = one column

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'Name': ['Ravi', 'Sreya', 'Ankit', 'Priya', 'Rekha'],
    'Age': [23, 25, 22, 24, 26],
    'City': ['Delhi', 'Kolkata', 'Mumbai', 'Chennai', 'Pune'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Female']
})

print(df)
```

```python
df['Name'] = df['Name'].map(lambda x: x.upper() if x.startswith('R') else x)
df
```

here lambda x, x is a single value just like when we apply df['col'].apply(lambda x)
same with .map
Here .map is a pandas function which automatically loops (iterates) over each element internally.
So here df['Name'].map, here map looks over each value in Name columns one by one and apply .upper function
